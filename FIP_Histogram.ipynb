```python
import matplotlib.pyplot as plt
Img=plt.imread('/content/drive/MyDrive/IMG-20220828-WA0001 (1).jpg')
```

```python
plt.imshow(Img)
```

```python
print(Img.shape)
```

```python
import cv2
Img1=cv2.cvtColor(Img, cv2.COLOR_BGR2GRAY)
#cvtcolor is used for converting color image to gray scale
```

```python
print(Img1.shape)
#only 1 channel is present now , grayscale
```

```python
Img2=Img1.ravel()
#ravel will convert the image into vectors
```

```python
Img2.shape
```

```python
plt.hist(Img2,256,[0,256])
plt.show()
```

```python
histogram=cv2.calcHist([Img1],[0],None,[256],[0,256])
plt.plot(histogram)
plt.show()
```

```python
Img3=cv2.equalizeHist(Img1)
```

```python
plt.hist(Img3.ravel(),  256,[0,256])
plt.show()
```

```python
plt.imshow(Img3,cmap='gray')
plt.show()
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2,2,1)
plt.imshow(Img)
plt.title('Original Image')
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.subplot(2,2,2)
plt.hist(Img1.ravel(),256,[0,256])
plt.title('Histogram of original image')
plt.xlabel('Gray levels')
plt.ylabel('Frequency of occurence')
plt.subplot(2,2,3)
plt.imshow(Img3,cmap='gray')
plt.title('Histogram Equalized image')
plt.subplot(2,2,4)
plt.hist(Img3.ravel(),256,[0,256])
plt.title('Equalized Histogram')
plt.show()
```
