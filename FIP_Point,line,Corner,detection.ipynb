```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/drive/MyDrive/carr.jpg",0)
```

```python
plt.imshow(img,cmap="gray")
```

```python
#create the kernel
kernel=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
kernel
```

```python
img_out=cv2.filter2D(img,-1,kernel)
```

```python
plt.imshow(img_out,cmap="gray")
```
