```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"/content/Screenshot 2023-08-20 235200.png",cv.IMREAD_GRAYSCALE)
```

```python
#Generate random Gaussian noise
mean=0
stddev=200
noise=np.zeros(img.shape,np.uint8)
cv2.randn(noise,mean,stddev)
```

```python
#adding noise to image
noisy_img=cv2.add(img,noise)
cv2.imwrite("Gauss_noisy_img.jpg",noisy_img)
```

```python
kernel=np.ones((5,5),np.float64)/25
dst=cv2.filter2D(noisy_img,-1,kernel)
```

```python
plt.subplot(121)
plt.imshow(noise,"gray")
plt.subplot(122)
plt.hist(noise.ravel(),256,[0,256])
plt.show()
```

```python
plt.subplot(231), plt.imshow(img,"gray")
plt.subplot(232), plt.imshow(noisy_img,"gray")
plt.subplot(233), plt.imshow(dst,"gray"), plt.title("Averaging")
plt.subplot(234), plt.hist(img.ravel(),256,[0,256]), plt.title("hist of original")
plt.subplot(235), plt.hist(noisy_img.ravel(),256,[0,256]), plt.title("hist of noisy")
plt.subplot(236), plt.hist(dst.ravel(),256,[0,256]), plt.title("hist of reconstructed")
plt.show()
```
